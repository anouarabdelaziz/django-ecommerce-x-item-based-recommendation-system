```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
tfid_inst = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0, stop_words='english')

tfid_matrix = tfid_inst.fit_transform(dff['title_description'])

cosine_similarity = linear_kernel(tfid_matrix, tfid_matrix)

result = {}
related= {}

for i,row in dff.iterrows():
    similar_indices = cosine_similarity[i].argsort()[:-50:-1] #getting the indices for the 50  most semilar items to the current item
    similar_items = [(cosine_similarity[i][j], dff.index[j]) for j in similar_indices]
    result[dff.index[i]+1] = similar_items[1:] #the first semilar item is the item it self no need to include it
    related[dff.index[i]+1] =[dff.index[j] for j in similar_indices] #getting the similar indices for a later use in django
    
    
    
```
